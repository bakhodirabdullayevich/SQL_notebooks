**Union all**

````python
jan = pd.DataFrame({'month':['jan','jan'],'revenue':[1,2]})
feb = pd.DataFrame({'month':['feb','feb'],'revenue':[1,2]})
````

````python
jan.to_sql('jan',con,index=False,if_exists='replace')
feb.to_sql('feb',con,index=False,if_exists='replace')
````

````python
sql = '''select * from jan t
union all
select * from feb t'''
````

````python
select(sql)
````

````python
sql = '''select t.revenue from jan t
union
select t.revenue from feb t'''
select(sql)
````

````python

sql = '''select t.revenue,
'jan' as month
 from jan t
union all
select t.revenue,
'mar' as month
 from feb t'''

select(sql)
````

# Maxsus funksiyalar
`substr()` funksiyasi berilgan matnning bir qismini (substring) ajratib olish uchun ishlatiladi. `substr(t.sex, 1, 1)` quyidagilarni anglatadi:

`t.sex`: Qaysi ustundan matn olinishi kerakligi (bu holatda` 'sex'` ustuni).

1: Matnni qaysi pozitsiyadan boshlab olish kerakligi (SQLda pozitsiyalar 1 dan boshlanadi).

1: Qancha miqdordagi belgilarni olish kerakligi.
Shunday qilib, bu kod sex ustunidagi har bir qiymatning birinchi belgisini qaytaradi.

````python
sql = '''select t.sex, substr(t.sex,1,1) from german_credit t'''
````

````python
select(sql)
````

SQLite'da mavjud emas. SQLite'da matnning bir qismini olish uchun faqat substr() funksiyasi ishlatiladi.

````python
sql = '''select t.sex, substring(t.sex,1,1), left(t.sex,1) from german_credit t'''
````

````python
select(sql)
````
