```python
import sqlite3
import pandas as pd
```

```python
con = sqlite3.connect('db')
```

```python
df = pd.DataFrame({'col':[1,2,3]})
```

```python
df
```

```python
df.to_sql('test_table', con, index=False, if_exists='replace')
```

```python
sql = '''select * from test_table t'''
```

```python
print(sql)
```

```python
pd.read_sql(sql, con)
```
